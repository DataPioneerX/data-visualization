```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Load the dataset
df=pd.read_csv(r"C:\Users\rehan shaikh\Desktop\my project.csv")



# Display the first few rows and summary statistics
df.sample(10)
df.describe(include="all")
```

```python

# Handle missing values
df.isnull().sum()
# there're missing value for analysis I'm considering the Technology and replacing the missing value with mode and change it to boolean function
mode_technology=df["Technology"].mode()[0]
mode_technology
df["Technology"]=df["Technology"].replace(np.nan,mode_technology)
df.dtypes

#after this code previously the technology is Object with 1 missing value i replaced the value with mode 

# Convert data types if necessary

df=pd.get_dummies(df,columns=["Technology"])####I've considered the Technology column from object I converted this to boolean and also added the new column for Technology
df.head()
df.dtypes



# Create any new columns needed for analysis
#created above a new Technology column
```
